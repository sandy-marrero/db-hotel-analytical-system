```python
import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted

class RoomDescription:
    def __init__(self, user, host):
        self.user = user
        self.host = host
        roomdescriptions =  requests.get(f"{host}/deceptecons/roomdescription").json()
        roomdescriptions.sort(key=lambda x: x['rdid'])
        roomdescriptions = list(map(lambda x: f"{x['rdid']}: {x['rname']}, {x['capacity']}, {x['rtype']}, {x['ishandicap']} ",roomdescriptions))
        self.create_roomDescription_rdid_combobox = widgets.Dropdown (
            placeholder='Choose a Room Description',
            options=roomdescriptions,
            description='Room Description:',
            ensure_option=True,
            disabled=False
        )
        self.update_roomDescription_rdid_combobox = widgets.Dropdown (
            placeholder='Choose a Room Description',
            options=roomdescriptions,
            description='Room Description:',
            ensure_option=True,
            disabled=False
        )
        self.delete_roomDescription_rdid_combobox = widgets.Dropdown (
            placeholder='Choose a Room Description',
            options=roomdescriptions,
            description='Room Description:',
            ensure_option=True,
            disabled=False
        )
        self.create_roomDescription_room_name_textInput = widgets.Text(
            placeholder='Enter room name',
            description='Room Name:',
            disabled=False
        )
        self.update_roomDescription_room_name_textInput = widgets.Text(
            placeholder='Enter room name',
            description='Room Name:',
            disabled=False
        )
        self.delete_roomDescription_room_name_textInput = widgets.Text(
            placeholder='Enter room name',
            description='Room Name:',
            disabled=False
        )
        self.create_roomDescription_room_type_dropdown = widgets.Dropdown(
            placeholder='Choose the type of room',
            options=["Basic", "Premium", "Deluxe", "Suite"],
            description='Room type:',
            ensure_option=True,
            disabled=False
        )
        self.update_roomDescription_room_type_dropdown = widgets.Dropdown(
            placeholder='Choose the type of room',
            options=["Basic", "Premium", "Deluxe", "Suite"],
            description='Room type:',
            ensure_option=True,
            disabled=False
        )
        self.delete_roomDescription_room_type_dropdown = widgets.Dropdown(
            placeholder='Choose the type of room',
            options=["Basic", "Premium", "Deluxe", "Suite"],
            description='Room type:',
            ensure_option=True,
            disabled=False
        )
        self.create_roomDescription_capacity_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=8,
            step=1,
            description='Capacity:',
            disabled=False
        )
        self.update_roomDescription_capacity_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=8,
            step=1,
            description='Capacity:',
            disabled=False
        )
        self.delete_roomDescription_capacity_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=8,
            step=1,
            description='Capacity:',
            disabled=False
        )
        self.create_roomDescription_handicap_checkbox = widgets.Checkbox(
            description='Handicap:',
            value=False
        )
        self.update_roomDescription_handicap_checkbox = widgets.Checkbox(
            description='Handicap:',
            value=False
        )
        self.delete_roomDescription_handicap_checkbox = widgets.Checkbox(
            description='Handicap:',
            value=False
        )
        self.submit_button_update =  widgets.Button(
            description='Update Room Description',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update Room Description',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete Room Description',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete Room Description',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_create =  widgets.Button(
            description='Create Room Description',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create Room Description',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        
    def submitUpdate(self,b):
        if self.update_roomDescription_room_name_textInput.value != "":
            rdid = self.update_roomDescription_rdid_combobox.value.split(':')[0]
            rname = self.update_roomDescription_room_name_textInput.value
            rtype = self.update_roomDescription_room_type_dropdown.value
            capacity = self.update_roomDescription_capacity_boundedfloattext.value
            isHandicap = self.update_roomDescription_handicap_checkbox.value
            data = {"rname": rname, "rtype": rtype, "capacity": capacity, "isHandicap": isHandicap}
            requests.put(f"{self.host}/deceptecons/roomdescription/{rdid}", json=data)

    def submitCreate(self,b):
            rname = self.create_roomDescription_room_name_textInput.value
            rtype = self.create_roomDescription_room_type_dropdown.value
            capacity = self.create_roomDescription_capacity_boundedfloattext.value
            isHandicap = self.create_roomDescription_handicap_checkbox.value
            data = {"rname": rname, "rtype": rtype, "capacity": capacity, "isHandicap": isHandicap}
            requests.post(f"{self.host}/deceptecons/roomdescription", json=data)

    
    def submitDelete(self,b):
        if self.delete_roomDescription_rdid_combobox.value != "":
            rdid = self.delete_roomDescription_rdid_combobox.value.split(':')[0]
            requests.delete(f"{host}/deceptecons/roomdescription/{rdid}")
        

    def displayRoomDescriptionDelete(self):
        display(widgets.HBox([
            self.delete_roomDescription_rdid_combobox,
            self.submit_button_delete
        ]))

    def displayRoomDescriptionCreate(self):
        display(widgets.HBox([
            self.create_roomDescription_room_name_textInput,
            self.create_roomDescription_room_type_dropdown,
            self.create_roomDescription_capacity_boundedfloattext,
            self.create_roomDescription_handicap_checkbox,
            self.submit_button_create
        ]))
    
    def displayRoomDescriptionUpdate(self):
        display(widgets.HBox([
            self.update_roomDescription_rdid_combobox,
            self.update_roomDescription_room_name_textInput,
            self.update_roomDescription_room_type_dropdown,
            self.update_roomDescription_capacity_boundedfloattext,
            self.update_roomDescription_handicap_checkbox,
            self.submit_button_update
        ]))
        
def run(user, host, output, purpose):
    x = RoomDescription(user, host)
    with output:
        if purpose == 'Update':
            x.displayRoomDescriptionUpdate()
        elif purpose == 'Create':
            x.displayRoomDescriptionCreate()
        elif purpose == 'Delete':
            x.displayRoomDescriptionDelete()
    
```
