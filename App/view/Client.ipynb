```python


import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted

class Client:
    def __init__(self, user, host):
        self.user = user
        self.host = host
        clients =  requests.get(f"{host}/deceptecons/client").json()
        clients.sort(key=lambda x: x['clid'])
        clients = list(map(lambda x: f"{x['clid']}: {x['fname']}, {x['lname']}, {x['age']}, {x['memberyear']} ",clients))
        
        self.create_client_clid_combobox = widgets.Dropdown (
            placeholder='Choose a Client',
            options=clients,
            description='Client:',
            ensure_option=True,
            disabled=False
        )
        self.update_client_clid_combobox = widgets.Dropdown (
            placeholder='Choose a Client',
            options=clients,
            description='Client:',
            ensure_option=True,
            disabled=False
        )
        self.delete_client_clid_combobox = widgets.Dropdown (
            placeholder='Choose a Client',
            options=clients,
            description='Client:',
            ensure_option=True,
            disabled=False
        )
        
        self.create_client_first_name_textInput = widgets.Text(
            placeholder='Enter first name',
            description='First Name:',
            disabled=False
        )
        self.update_client_first_name_textInput = widgets.Text(
            placeholder='Enter first name',
            description='First Name:',
            disabled=False
        )
        self.delete_client_first_name_textInput = widgets.Text(
            placeholder='Enter first name',
            description='First Name:',
            disabled=False
        )
        
        self.create_client_last_name_textInput = widgets.Text(
            placeholder='Enter last name',
            description='Last Name:',
            disabled=False
        )
        self.update_client_last_name_textInput = widgets.Text(
            placeholder='Enter last name',
            description='Last Name:',
            disabled=False
        )
        self.delete_client_last_name_textInput = widgets.Text(
            placeholder='Enter last name',
            description='Last Name:',
            disabled=False
        )
        
        self.create_client_age_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=115,
            step=1,
            description='Age:',
            disabled=False
        )
        self.update_client_age_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=115,
            step=1,
            description='Age:',
            disabled=False
        )
        self.delete_client_age_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=115,
            step=1,
            description='Age:',
            disabled=False
        )
        
        self.create_client_memberyear_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000,
            step=1,
            description='Member year:',
            disabled=False
        )
        self.update_client_memberyear_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000,
            step=1,
            description='Member year:',
            disabled=False
        )
        self.delete_client_memberyear_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000,
            step=1,
            description='Member year:',
            disabled=False
        )
    
        self.submit_button_update =  widgets.Button(
            description='Update Client',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update Client',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete Client',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete Room Description',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_create =  widgets.Button(
            description='Create Client',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create Client',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        
    def submitUpdate(self,b):
        if self.update_client_clid_combobox.value != "":
            clid = self.update_client_clid_combobox.value.split(':')[0]
            fname = self.update_client_first_name_textInput.value
            lname = self.update_client_last_name_textInput.value
            age = self.update_client_age_boundedfloattext.value
            memberyear = self.update_client_memberyear_boundedfloattext.value
            data = {"fname": fname, "lname": lname, "age": age, "memberyear": memberyear}
            requests.put(f"{self.host}/deceptecons/client/{clid}", json=data)

    def submitCreate(self,b):
            fname = self.create_client_first_name_textInput.value
            lname = self.create_client_last_name_textInput.value
            age = self.create_client_age_boundedfloattext.value
            memberyear = self.create_client_memberyear_boundedfloattext.value
            data = {"fname": fname, "lname": lname, "age": age, "memberyear": memberyear}
            requests.post(f"{self.host}/deceptecons/client", json=data)

    
    def submitDelete(self,b):
        if self.delete_client_clid_combobox.value != "":
            clid = self.delete_client_clid_combobox.value.split(':')[0]
            requests.delete(f"{self.host}/deceptecons/client/{clid}")
        

    def displayClientDelete(self):
        display(widgets.VBox([
            self.delete_client_clid_combobox,
            self.submit_button_delete
        ]))

    def displayClientCreate(self):
        display(widgets.VBox([
            self.create_client_first_name_textInput,
            self.create_client_last_name_textInput,
            self.create_client_age_boundedfloattext,
            self.create_client_memberyear_boundedfloattext,
            self.submit_button_create
        ]))
    
    def displayClientUpdate(self):
        display(widgets.VBox([
            self.update_client_clid_combobox,
            self.update_client_first_name_textInput,
            self.update_client_last_name_textInput,
            self.update_client_age_boundedfloattext,
            self.update_client_memberyear_boundedfloattext,
            self.submit_button_update
        ]))
        
def run(user, host, output, purpose):
    x = Client(user, host)
    with output:
        if purpose == 'Update':
            x.displayClientUpdate()
        elif purpose == 'Create':
            x.displayClientCreate()
        elif purpose == 'Delete':
            x.displayClientDelete() 
```
