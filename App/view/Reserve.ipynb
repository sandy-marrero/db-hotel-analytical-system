```python
import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted
from datetime import datetime

class Reserve:
    def __init__(self, user, host, output):
        self.user = user
        self.host = host
        self.out = output
        reservations = requests.get(f"{host}/deceptecons/reserve").json()
        natsorted(reservations)
        reservations = list(map(lambda x: f"{x['reid']}", reservations))
        self.update_reserve_reid_dropdown = widgets.Dropdown (
            placeholder='Choose a Reservation ID',
            options=reservations,
            description='Reservation ID:',
            ensure_option=True,
            disabled=False
        )
        self.delete_reserve_reid_dropdown = widgets.Dropdown (
            placeholder='Choose a Reservation ID',
            options=reservations,
            description='Reservation ID:',
            ensure_option=True,
            disabled=False
        )
        self.create_reserve_reid_dropdown = widgets.Dropdown (
            placeholder='Choose a Reservation ID',
            options=reservations,
            description='Reservation ID:',
            ensure_option=True,
            disabled=False
        )
        roomunavailables = requests.get(f"{host}/deceptecons/roomunavailable").json()
        roomunavailables.sort(key=lambda x: x['ruid'])
        roomunavailables = list(map(lambda x: f"{x['ruid']}", roomunavailables))
        self.update_reserve_ruid_dropdown = widgets.Dropdown (
            placeholder='Choose a RoomUnavailable ID:',
            options=roomunavailables,
            description='Room Unvailable ID:',
            ensure_option=True,
            disabled=False
        )
        self.create_reserve_ruid_dropdown = widgets.Dropdown (
            placeholder='Choose a RoomUnavailable ID:',
            options=roomunavailables,
            description='Room Unvailable ID:',
            ensure_option=True,
            disabled=False
        )
        self.delete_reserve_ruid_dropdown = widgets.Dropdown (
            placeholder='Choose a RoomUnavailable ID:',
            options=roomunavailables,
            description='Room Unvailable ID:',
            ensure_option=True,
            disabled=False
        )

        clients = requests.get(f"{host}/deceptecons/client").json()
        clients.sort(key=lambda x: x['clid'])
        clients = list(map(lambda x: x['clid'],clients))

        self.create_reserve_clid_dropdown = widgets.Dropdown (
            placeholder='Choose a Client ID:',
            options=clients,
            description='Client ID:',
            ensure_option=True,
            disabled=False
        )
        
        self.update_reserve_clid_dropdown = widgets.Dropdown (
            placeholder='Choose a Client ID:',
            options=clients,
            description='Client ID:',
            ensure_option=True,
            disabled=False
        )
        
        self.delete_reserve_clid_dropdown = widgets.Dropdown (
            placeholder='Choose a Client ID:',
            options=clients,
            description='Client ID:',
            ensure_option=True,
            disabled=False
        )

        self.create_total_cost_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Reservation Total Cost:',
            disabled=False)
        
        self.update_total_cost_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Reservation Total Cost:',
            disabled=False)
        
        self.delete_total_cost_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Reservation Total Cost:',
            disabled=False)

        self.create_payment_dropdown = widgets.Dropdown (
            placeholder='Choose a Payment Method:',
            options=['cash', 'credit card', 'pear pay', 'debit card', 'check'],
            description='Payment Method:',
            ensure_option=True,
            disabled=False
        )
        self.update_payment_dropdown = widgets.Dropdown (
            placeholder='Choose a Payment Method:',
            options=['cash', 'credit card', 'pear pay', 'debit card', 'check'],
            description='Payment Method:',
            ensure_option=True,
            disabled=False
        )
        self.delete_payment_dropdown = widgets.Dropdown (
            placeholder='Choose a Payment Method:',
            options=['cash', 'credit card', 'pear pay', 'debit card', 'check'],
            description='Payment Method:',
            ensure_option=True,
            disabled=False
        )

        self.create_guests_boundedinttext = widgets.BoundedIntText(
            min=1,
            max=10000,
            step=1,
            description='Guests:',
            disabled=False
        )
        self.update_guests_boundedinttext = widgets.BoundedIntText(
            min=1,
            max=10000,
            step=1,
            description='Guests:',
            disabled=False
        )
        self.delete_guests_boundedinttext = widgets.BoundedIntText(
            min=1,
            max=10000,
            step=1,
            description='Guests:',
            disabled=False
        )
        
        self.submit_button_update =  widgets.Button(
            description='Update Reservation',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update Reservation',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        
        self.submit_button_create =  widgets.Button(
            description='Create Reservation',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create Reservation',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete Reservation',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete Reservation',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.create_roomunavailable_start_date = widgets.DatePicker(description='Pick a Reservation Start Date')
        self.create_roomunavailable_end_date = widgets.DatePicker(description='Pick a Reservation Start Date')

        rooms = requests.get(f"{self.host}/deceptecons/room").json()
        rooms = list(map(lambda x: x['rid'], rooms))
        rooms.sort()

        self.create_roomunavailable_rid = widgets.Dropdown(
        options=rooms,
        description='Room ID:',
        disabled=False,) 
        
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        

    def submitUpdate(self,b):
        if self.update_reserve_reid_dropdow.value != "" and \
        self.update_reserve_ruid_dropdown.value != "" and \
        self.update_reserve_clid_dropdown.value is not None and \
        self.update_total_cost_boundedfloattext.value is not None and \
        self.update_payment_dropdown.value is not None and \
        self.update_guests_boundedinttext.value is not None:
            reid = self.update_reserve_reid_dropdown.value
            ruid = self.update_reserve_ruid_dropdown.value
            clid = self.update_reserve_clid_dropdown.value
            total_cost = self.update_total_cost_boundedfloattext.value
            payment = self.update_payment_dropdown.value
            guests = self.update_guests_boundedinttext.value
            data = {"reid": reid, "ruid": ruid, 'clid': clid, "total_cost": total_cost, 'payment': payment, 'guests': guests}
            requests.put(f"{host}/deceptecons/reserve/{reid}", json=data)

    def submitCreate(self, b):
        if self.create_roomunavailable_start_date.value is not None and \
        self.create_roomunavailable_end_date.value is not None and \
        self.create_roomunavailable_rid.value is not None and\
        self.create_reserve_clid_dropdown.value is not None and \
        self.create_payment_dropdown.value is not None and \
        self.create_guests_boundedinttext.value is not None:
            startdate = self.create_roomunavailable_start_date.value
            enddate = self.create_roomunavailable_end_date.value
            temp = None
            if enddate < startdate:
                temp = startdate
                startdate = enddate
                enddate = temp
            startdate = str(startdate)
            enddate = str(enddate)
            rid = self.create_roomunavailable_rid.value
            data = {"startdate": startdate, "enddate": enddate, "rid": rid}
            ruid = int(requests.post(f"{self.host}/deceptecons/roomunavailable", json=data).json())
            clid = self.create_reserve_clid_dropdown.value
            payment = self.create_payment_dropdown.value
            guests = self.create_guests_boundedinttext.value
            data = {"ruid": ruid, 'clid': clid, 'payment': payment, 'guests': guests}
            requests.post(f"{host}/deceptecons/reserve", json=data)
            with self.out:
                print("Reservation made succesfully")

    def submitDelete(self,b):
        if self.delete_reserve_reid_dropdown.value != "":
            reid = self.delete_reserve_reid_dropdown.value
            requests.delete(f"{host}/deceptecons/reserve/{reid}")

    def displayReserveUpdate(self):
        display(widgets.VBox([
            self.update_reserve_reid_dropdown,
            self.update_reserve_ruid_dropdown,
            self.update_reserve_clid_dropdown,
            self.update_total_cost_boundedfloattext,
            self.update_payment_dropdown,
            self.update_guests_boundedinttext,
            self.submit_button_update
        ]))

    def displayReserveCreate(self):
        display(
            widgets.VBox([
                self.create_roomunavailable_start_date,
                self.create_roomunavailable_end_date,
                self.create_roomunavailable_rid,
                self.create_reserve_clid_dropdown,
                self.create_payment_dropdown,
                self.create_guests_boundedinttext,
                self.submit_button_create
            ])
        )

    def displayReserveDelete(self):
        display(
            widgets.VBox([
                self.delete_reserve_reid_dropdown,
                self.submit_button_delete
            ])
        )
            

def run(user, host, output, purpose):
    x = Reserve(user, host, output)
    with x.out:
        if purpose == 'Update':
            x.displayReserveUpdate()
        elif purpose == 'Create':
            x.displayReserveCreate()
        elif purpose == 'Delete':
            x.displayReserveDelete()
            
        
```
