```python
import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted

class Employee:
    def __init__(self, user, host):
        self.user = user
        self.host = host
        hotels =  requests.get(f"{self.host}/deceptecons/hotel").json()
        hotels.sort(key=lambda x: x['hid'])
        hotels = list(map(lambda x: f"{x['hid']}: {x['hname']}",filter(lambda x: x['chid'] != -1 and x['hid'] != -1,hotels)))
        self.create_hotel_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        self.update_hotel_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        self.dalete_room_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        employees = requests.get(f"{host}/deceptecons/employee").json()
        employees.sort(key=lambda x: x['eid'])
        employees = list(map(lambda x: str(x['eid']),employees))
        natsorted(employees)
        self.create_employee_eid_combobox = widgets.Dropdown (
            placeholder='Choose an Employee',
            options=employees,
            description='Employee ID:',
            ensure_option=True,
            disabled=False
        )
        self.update_employee_eid_combobox = widgets.Dropdown (
            placeholder='Choose an Employee',
            options=employees,
            description='Employee ID:',
            ensure_option=True,
            disabled=False
        )
        self.delete_employee_eid_combobox = widgets.Dropdown (
            placeholder='Choose an Employee',
            options=employees,
            description='Employee ID:',
            ensure_option=True,
            disabled=False
        )
        self.create_employee_first_name_textInput = widgets.Text(
            placeholder='Enter first name',
            description='First Name:',
            disabled=False
        )
        self.update_employee_first_name_textInput = widgets.Text(
            placeholder='Enter first name',
            description='First Name:',
            disabled=False
        )
        self.delete_employee_first_name_textInput = widgets.Text(
            placeholder='Enter first name',
            description='First Name:',
            disabled=False
        )
        self.create_employee_last_name_textInput = widgets.Text(
            placeholder='Enter last name',
            description='Last Name:',
            disabled=False
        )
        self.update_employee_last_name_textInput = widgets.Text(
            placeholder='Enter last name',
            description='Last Name:',
            disabled=False
        )
        self.delete_employee_last_name_textInput = widgets.Text(
            placeholder='Enter last name',
            description='Last Name:',
            disabled=False
        )
        self.create_employee_age_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=200,
            step=1,
            description='Age:',
            disabled=False
        )
        self.update_employee_age_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=200,
            step=1,
            description='Age:',
            disabled=False
        )
        self.delete_employee_age_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=200,
            step=1,
            description='Age:',
            disabled=False
        )
        self.create_employee_position_dropdown = widgets.Dropdown(
            placeholder='Choose the employee position',
            options=["Regular", "Supervisor", "Administrator"],
            description='Employee position:',
            ensure_option=True,
            disabled=False
        )
        self.update_employee_position_dropdown = widgets.Dropdown(
            placeholder='Choose the employee position',
            options=["Regular", "Supervisor", "Administrator"],
            description='Employee position:',
            ensure_option=True,
            disabled=False
        )
        self.delete_employee_position_dropdown = widgets.Dropdown(
            placeholder='Choose the employee position',
            options=["Regular", "Supervisor", "Administrator"],
            description='Employee position:',
            ensure_option=True,
            disabled=False
        )
        self.create_employee_salary_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Employee Salary:',
            disabled=False
        )
        self.update_employee_salary_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Employee Salary:',
            disabled=False
        )
        self.delete_employee_salary_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Employee Salary:',
            disabled=False
        )
        self.submit_button_update =  widgets.Button(
            description='Update Employee',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update Employee',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete Employee',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete Employee',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_create =  widgets.Button(
            description='Create Employee',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create Employee',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        
    def submitUpdate(self,b):
        if (self.update_employee_eid_combobox.value != "" and \
        self.update_hotel_hid_combobox.value != "") == True:
            eid = self.update_employee_eid_combobox.value
            hid = self.update_hotel_hid_combobox.value.split(':')[0]
            fname = self.update_employee_first_name_textInput.value
            lname = self.update_employee_last_name_textInput.value
            age = self.update_employee_age_boundedfloattext.value
            position = self.update_employee_position_dropdown.value
            salary = self.update_employee_salary_boundedfloattext.value
            data = {"hid": hid, "fname": fname, "lname": lname, "age": age, "position": position, "salary": salary}
            requests.put(f"{self.host}/deceptecons/employee/{eid}", json=data)

    def submitCreate(self,b):
        if self.create_hotel_hid_combobox.value != "":
            hid = self.create_hotel_hid_combobox.value.split(':')[0]
            fname = self.create_employee_first_name_textInput.value
            lname = self.create_employee_last_name_textInput.value
            age = self.create_employee_age_boundedfloattext.value
            position = self.create_employee_position_dropdown.value
            salary = self.create_employee_salary_boundedfloattext.value
            data = {"hid": hid, "fname": fname, "lname": lname, "age": age, "position": position, "salary": salary}
            requests.post(f"{self.host}/deceptecons/employee", json=data)

    
    def submitDelete(self,b):
        if self.delete_employee_eid_combobox.value != "":
            eid = self.delete_employee_eid_combobox.value.split(':')[0]
            requests.delete(f"{host}/deceptecons/employee/{eid}")
        

    def displayEmployeeDelete(self):
        display(widgets.VBox([
            self.delete_employee_eid_combobox,
            self.submit_button_delete
        ]))

    def displayEmployeeCreate(self):
        display(widgets.VBox([
            self.create_hotel_hid_combobox,
            self.create_employee_first_name_textInput,
            self.create_employee_last_name_textInput,
            self.create_employee_age_boundedfloattext,
            self.create_employee_position_dropdown,
            self.create_employee_salary_boundedfloattext,
            self.submit_button_create
        ]))
    
    def displayEmployeeUpdate(self):
        display(widgets.VBox([
            self.update_employee_eid_combobox,
            self.update_hotel_hid_combobox,
            self.update_employee_first_name_textInput,
            self.update_employee_last_name_textInput,
            self.update_employee_age_boundedfloattext,
            self.update_employee_position_dropdown,
            self.update_employee_salary_boundedfloattext,
            self.submit_button_update
        ]))
        
def run(user, host, output, purpose):
    x = Employee(user, host)
    with output:
        if purpose == 'Update':
            x.displayEmployeeUpdate()
        elif purpose == 'Create':
            x.displayEmployeeCreate()
        elif purpose == 'Delete':
            x.displayEmployeeDelete()
    
```
