```python

import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted

class Hotel:
    def __init__(self, user, host):
        self.user = user
        self.host = host
        chains =  requests.get(f"{self.host}/deceptecons/chains").json()
        chains.sort(key=lambda x: x['chid'])
        # chains = list(map(lambda x: f"{x['chid']}: {x['cname']}",filter(lambda x: x['chid'] != -1,chains)))
        chains = [f"{chain['chid']}: {chain['name']}" for chain in chains]
        
        self.create_chain_chid_combobox = widgets.Dropdown (
            placeholder='Choose a Chain',
            options=chains,
            description='Chain:',
            ensure_option=True,
            disabled=False
        )
        self.update_chain_chid_combobox = widgets.Dropdown (
            placeholder='Choose a Chain',
            options=chains,
            description='Chain:',
            ensure_option=True,
            disabled=False
        )
        self.delete_chains_chid_combobox = widgets.Dropdown (
            placeholder='Choose a chains',
            options=chains,
            description='Chain:',
            ensure_option=True,
            disabled=False
        )
        
        hotels =  requests.get(f"{host}/deceptecons/hotel").json()
        hotels.sort(key=lambda x: x['hid'])
        hotels = list(map(lambda x: f"{x['hid']}: {x['chid']}, {x['hname']}, {x['hcity']}",hotels))
        
        self.create_hotel_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        self.update_hotel_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        self.delete_hotel_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )

        self.create_hotel_name_textInput = widgets.Text(
            placeholder = "Enter hotel name:",
            description ='Hotel name:',
            disabled=False
        )    
        self.create_hotel_city_textInput = widgets.Text(
            placeholder = "Enter city name",
            description='City name:',
            disabled=False
        )    
        
        self.update_hotel_name_textInput = widgets.Text(
            placeholder = "Enter hotel name:",
            description ='Hotel name:',
            disabled=False
        )   
        self.update_hotel_city_textInput = widgets.Text(
            placeholder = "Enter city name",
            description='City name:',
            disabled=False
        )
        
        self.delete_hotel_name_textInput = widgets.Text(
            placeholder = "Enter hotel name:",
            description ='Hotel name:',
            disabled=False
        )  
        self.delete_hotel_city_textInput = widgets.Text(
            placeholder = "Enter city name",
            description='City name:',
            disabled=False
        ) 
        
        self.submit_button_update =  widgets.Button(
            description='Update Hotel',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update Hotel',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete Hotel',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete Hotel',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_create =  widgets.Button(
            description='Create Hotel',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create Hotel',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        
    def submitUpdate(self,b):
        if (self.update_hotel_hid_combobox.value != "" and \
        self.update_chain_chid_combobox.value != "") == True:
            hid = self.update_hotel_hid_combobox.value.split(':')[0]
            chid = self.update_chain_chid_combobox.value.split(':')[0]
            hname = self.update_hotel_name_textInput.value
            hcity = self.update_hotel_city_textInput.value
            data = {"chid": chid, "hname": hname, "hcity": hcity}
            print(data)
            requests.put(f"{self.host}/deceptecons/hotel/{hid}", json=data)

    def submitCreate(self,b):
        if (self.create_hotel_hid_combobox.value != "" and \
        self.create_chain_chid_combobox != "") == True:
            chid = self.create_chain_chid_combobox.value.split(':')[0]
            hname = self.create_hotel_name_textInput.value
            hcity = self.create_hotel_city_textInput.value
            data = {"chid": chid, "hname": hname, "hcity": hcity}
            requests.post(f"{self.host}/deceptecons/hotel", json=data)

    
    def submitDelete(self,b):
        if self.delete_hotel_hid_combobox.value != "":
            hid = self.delete_hotel_hid_combobox.value.split(':')[0]
            requests.delete(f"{self.host}/deceptecons/hotel/{hid}")
        

    def displayHotelDelete(self):
        display(widgets.VBox([
            self.delete_hotel_hid_combobox,
            self.submit_button_delete
        ]))

    def displayHotelCreate(self):
        display(widgets.VBox([
            self.create_chain_chid_combobox,
            self.create_hotel_name_textInput,
            self.create_hotel_city_textInput,
            self.submit_button_create
        ]))
    
    def displayHotelUpdate(self):
        display(widgets.VBox([
            self.update_hotel_hid_combobox,
            self.update_chain_chid_combobox,
            self.update_hotel_name_textInput,
            self.update_hotel_city_textInput,
            self.submit_button_update
        ]))
        
def run(user, host, output, purpose):
    x = Hotel(user, host)
    with output:
        if purpose == 'Update':
            x.displayHotelUpdate()
        elif purpose == 'Create':
            x.displayHotelCreate()
        elif purpose == 'Delete':
            x.displayHotelDelete()
```
