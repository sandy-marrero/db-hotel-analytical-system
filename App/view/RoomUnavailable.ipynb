```python


import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted

class RoomUnavailable:
    def __init__(self, user, host):
        self.user = user
        self.host = host
        rooms =  requests.get(f"{self.host}/deceptecons/room").json()
        rooms.sort(key=lambda x: x['hid'])
        rooms = list(map(lambda x: f"{x['rid']}:",filter(lambda x: x['rid'] != -1 and x['hid'] != -1 and x['hid'] != -1,rooms)))
        self.create_room_rid_combobox = widgets.Dropdown (
            placeholder='Choose a Room',
            options=rooms,
            description='Room:',
            ensure_option=True,
            disabled=False
        )
        self.update_room_rid_combobox = widgets.Dropdown (
            placeholder='Choose a Room',
            options=rooms,
            description='Room:',
            ensure_option=True,
            disabled=False
        )
        self.delete_room_rid_combobox = widgets.Dropdown (
            placeholder='Choose a Room',
            options=rooms,
            description='Room:',
            ensure_option=True,
            disabled=False
        )
        
        roomunavailables = requests.get(f"{host}/deceptecons/roomunavailable").json()
        roomunavailables.sort(key=lambda x: x['ruid'])
        roomunavailables = list(map(lambda x: str(x['ruid']),roomunavailables))
        natsorted(roomunavailables)
        self.create_roomunavailable_ruid_combobox = widgets.Dropdown (
            placeholder='Choose an RoomUnavailable',
            options=roomunavailables,
            description='RoomUnavailable ID:',
            ensure_option=True,
            disabled=False
        )
        self.update_roomunavailable_ruid_combobox = widgets.Dropdown (
            placeholder='Choose an RoomUnavailable',
            options=roomunavailables,
            description='RoomUnavailable ID:',
            ensure_option=True,
            disabled=False
        )
        self.delete_roomunavailable_ruid_combobox = widgets.Dropdown (
            placeholder='Choose an RoomUnavailable',
            options=roomunavailables,
            description='RoomUnavailable ID:',
            ensure_option=True,
            disabled=False
        )
        
        self.update_roomunavavilable_startdate_datepicker = widgets.DatePicker (
            description='Choose a start date',
            disabled=False
        )
        self.create_roomunavavilable_startdate_datepicker = widgets.DatePicker (
            description='Choose a start date',
            disabled=False
        )
        self.delete_roomunavavilable_startdate_datepicker = widgets.DatePicker (
            description='Choose a start date',
            disabled=False
        )
        self.update_roomunavavilable_enddate_datepicker = widgets.DatePicker (
            description='Choose a end date',
            disabled=False
        )
        self.create_roomunavavilable_enddate_datepicker = widgets.DatePicker (
            description='Choose a end date',
            disabled=False
        )
        self.delete_roomunavavilable_enddate_datepicker = widgets.DatePicker (
            description='Choose a end date',
            disabled=False
        )
        self.submit_button_update =  widgets.Button(
            description='Update RoomUnavailable',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update RoomUnavailable',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete RoomUnavailable',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete RoomUnavailable',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_create =  widgets.Button(
            description='Create RoomUnavailable',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create RoomUnavailable',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        
    def submitUpdate(self,b):
        if (self.update_roomunavailable_ruid_combobox.value != "" and \
        self.update_room_rid_combobox.value != "" and \
        self.update_roomunavavilable_startdate_datepicker.value is not None and \
        self.update_roomunavavilable_startdate_datepicker.value is not None) == True:
            ruid = self.update_roomunavailable_ruid_combobox.value
            rid = self.update_room_rid_combobox.value.split(':')[0]
            startdate = self.update_roomunavavilable_startdate_datepicker.value
            enddate = self.update_roomunavavilable_enddate_datepicker.value
            data = {"rid": rid, "startdate": startdate, "enddate": enddate}
            requests.put(f"{self.host}/deceptecons/roomunavailable/{ruid}", json=data)

    def submitCreate(self,b):
        if self.create_room_rid_combobox.value != "":
            rid = self.update_room_rid_combobox.value.split(':')[0]
            startdate = self.create_roomunavavilable_startdate_datepicker
            enddate = self.create_roomunavavilable_enddate_datepicker
            data = {"rid": rid, "startdate": startdate, "enddate": enddate}
            requests.put(f"{self.host}/deceptecons/roomunavailable/", json=data)

    
    def submitDelete(self,b):
        if self.delete_roomunavailable_ruid_combobox.value != "":
            ruid = self.delete_roomunavailable_ruid_combobox.value.split(':')[0]
            requests.delete(f"{self.host}/deceptecons/roomunavailable/{ruid}")
        

    def displayRoomUnavailableDelete(self):
        display(widgets.VBox([
            self.delete_roomunavailable_ruid_combobox,
            self.submit_button_delete
        ]))

    def displayRoomUnavailableCreate(self):
        display(widgets.VBox([
            self.create_room_rid_combobox,
            self.create_roomunavavilable_startdate_datepicker,
            self.create_roomunavavilable_enddate_datepicker,
            self.submit_button_create
        ]))
    
    def displayRoomUnavailableUpdate(self):
        display(widgets.VBox([
            self.update_roomunavailable_ruid_combobox,
            self.update_room_rid_combobox,
            self.update_roomunavavilable_startdate_datepicker,
            self.update_roomunavavilable_enddate_datepicker,
            self.submit_button_update
        ]))
        
def run(user, host, output, purpose):
    x = RoomUnavailable(user, host)
    with output:
        if purpose == 'Update':
            x.displayRoomUnavailableUpdate()
        elif purpose == 'Create':
            x.displayRoomUnavailableCreate()
        elif purpose == 'Delete':
            x.displayRoomUnavailableDelete()
    

     
```
