```python
import ipywidgets as widgets
from IPython.display import display
import requests
from natsort import natsorted

class Room:
    def __init__(self, user, host, output):
        self.user = user
        self.host = host
        self.out = output
        hotels =  requests.get(f"{self.host}/deceptecons/hotel").json()
        hotels.sort(key=lambda x: x['hid'])
        hotels = list(map(lambda x: f"{x['hid']}: {x['hname']}",filter(lambda x: x['chid'] != -1 and x['hid'] != -1,hotels)))
        self.create_room_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        self.update_room_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        self.dalete_room_hid_combobox = widgets.Dropdown (
            placeholder='Choose a Hotel',
            options=hotels,
            description='Hotel:',
            ensure_option=True,
            disabled=False
        )
        roomdescriptions =  requests.get(f"{host}/deceptecons/roomdescription").json()
        roomdescriptions.sort(key=lambda x: x['rdid'])
        roomdescriptions = list(map(lambda x: f"{x['rdid']}: {x['rname']}, {x['capacity']}, {x['rtype']}, {x['ishandicap']} ",roomdescriptions))
        self.create_room_rdid_combobox = widgets.Dropdown (
            placeholder='Choose a Room Description',
            options=roomdescriptions,
            description='Room Description:',
            ensure_option=True,
            disabled=False
        )
        self.update_room_rdid_combobox = widgets.Dropdown (
            placeholder='Choose a Room Description',
            options=roomdescriptions,
            description='Room Description:',
            ensure_option=True,
            disabled=False
        )
        self.delete_room_rdid_combobox = widgets.Dropdown (
            placeholder='Choose a Room Description',
            options=roomdescriptions,
            description='Room Description:',
            ensure_option=True,
            disabled=False
        )
        rooms = requests.get(f"{host}/deceptecons/room").json()
        rooms.sort(key=lambda x: x['rid'])
        rooms = list(map(lambda x: str(x['rid']),rooms))
        natsorted(rooms)
        self.create_room_rid_combobox = widgets.Dropdown (
            placeholder='Choose a Room',
            options=rooms,
            description='Room ID:',
            ensure_option=True,
            disabled=False
        )
        self.update_room_rid_combobox = widgets.Dropdown (
            placeholder='Choose a Room',
            options=rooms,
            description='Room ID:',
            ensure_option=True,
            disabled=False
        )
        self.delete_room_rid_combobox = widgets.Dropdown (
            placeholder='Choose a Room',
            options=rooms,
            description='Room ID:',
            ensure_option=True,
            disabled=False
        )
        self.create_room_price_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Room Price:',
            disabled=False
        )
        self.update_room_price_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Room Price:',
            disabled=False
        )
        self.delete_room_price_boundedfloattext = widgets.BoundedFloatText(
            min=0,
            max=1000000000,
            step=0.01,
            description='Room Price:',
            disabled=False
        )
        self.submit_button_update =  widgets.Button(
            description='Update Room',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Update Room',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_delete =  widgets.Button(
            description='Delete Room',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Delete Room',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_create =  widgets.Button(
            description='Create Room',
            disabled=False,
            button_style='success', # 'success', 'info', 'warning', 'danger' or ''
            tooltip='Create Room',
            icon='check' # (FontAwesome names without the `fa-` prefix) 
        )
        self.submit_button_update.on_click(self.submitUpdate)
        self.submit_button_create.on_click(self.submitCreate)
        self.submit_button_delete.on_click(self.submitDelete)
        
    def submitUpdate(self,b):
        if (self.update_room_rid_combobox.value != "" and \
        self.update_room_hid_combobox.value != "" and \
        self.update_room_rdid_combobox != "") == True:
            rid = self.update_room_rid_combobox.value
            hid = self.update_room_hid_combobox.value.split(':')[0]
            rdid = self.update_room_rdid_combobox.value.split(':')[0]
            rprice = self.update_room_price_boundedfloattext.value
            data = {"hid": hid, "rdid": rdid, "rprice": rprice}
            requests.put(f"{self.host}/deceptecons/room/{rid}", json=data)

    def submitCreate(self,b):
        if (self.create_room_hid_combobox.value != "" and \
        self.create_room_rdid_combobox != "") == True:
            hid = self.create_room_hid_combobox.value.split(':')[0]
            rdid = self.create_room_rdid_combobox.value.split(':')[0]
            rprice = self.create_room_price_boundedfloattext.value
            data = {"hid": hid, "rdid": rdid, "rprice": rprice}
            requests.post(f"{self.host}/deceptecons/room", json=data)
            with self.out:
                print("Room Created Succesfully")

    
    def submitDelete(self,b):
        if self.delete_room_rid_combobox.value != "":
            hid = self.delete_room_rid_combobox.value.split(':')[0]
            requests.delete(f"{host}/deceptecons/room/{hid}")
        

    def displayRoomDelete(self):
        display(widgets.VBox([
            self.delete_room_rid_combobox,
            self.submit_button_delete
        ]))

    def displayRoomCreate(self):
        display(widgets.VBox([
            self.create_room_hid_combobox,
            self.create_room_rdid_combobox,
            self.create_room_price_boundedfloattext,
            self.submit_button_create
        ]))
    
    def displayRoomUpdate(self):
        display(widgets.VBox([
            self.update_room_rid_combobox,
            self.update_room_hid_combobox,
            self.update_room_rdid_combobox,
            self.update_room_price_boundedfloattext,
            self.submit_button_update
        ]))
        
def run(user, host, output, purpose):
    x = Room(user, host, output)
    with output:
        if purpose == 'Update':
            x.displayRoomUpdate()
        elif purpose == 'Create':
            x.displayRoomCreate()
        elif purpose == 'Delete':
            x.displayRoomDelete()
    
```
